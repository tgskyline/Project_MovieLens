**Movielens**

O conjunto de dados MovieLens IM contém um milhão de
avaliacões coletadas de 6 mil usuários sobre 4 mil filmes. Os dados estão espalhados em três tabelas: avaliações, informações de usuários e informações sobre filmes

[MovieLens](https://grouplens.org/datasets/movielens/)

```python
import pandas as pd

pd.options.display.max_rows = 10

unames = ['user_id', 'gender', 'age', 'occupation', 'zip']
user = pd.read_table('users.dat', sep='::', header=None, names=unames)

rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table('ratings.dat', sep='::', header=None, names=rnames)

mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table('movies.dat', sep='::', header=None, names=mnames)
```

```python
ratings[:5]
```

```python
movies[:5]
```

```python
ratings.info()
```

```python
ratings.isnull().sum()
user.isnull().sum()
movies.isnull().sum()
```

Analisar os dados espalhados em três tabelas não é uma tarefa simples; por exemplo, suponha que você quisesse calcular as avaliações médias
para um filme em particular, de acordo com o sexo e a idade (leia o arquivos README para entender as codificações dos dados).

Usando a função Merge do pandas, inicialmente faremos o merge de ratings e users, e então faremos o merge desse resultado com os dados de Movies.

Observe que o Pandas infere quais colunas deve usar como as chaves do merge com base nos nomes iguais de colunas

```python
data = pd.merge(pd.merge(ratings, user), movies)
data
```

Vamos calcular as avaliações médias de cada filme agrupadas pela identificação de gênero da pessoa

```python
mean_ratings = data.pivot_table('rating', index='title', columns='gender', aggfunc='mean')
mean_ratings[:20]
```

Vamos agrupar os dados por título e obter uma Series com os tamanhos dos grupos (quantidade de avaliações) para cada título:

```python
rating_by_title = data.groupby('title').size()
rating_by_title[:10]
```

Agora iremos exibir apenas os filmes com mais de 250 avaliações na base

```python
active_titles = rating_by_title.index[rating_by_title >= 3000]
active_titles
```

Para ver os principais filmes entre as telespectadoras femininas, podemos ordenar de acordo com a coluna F em ordem decrescente: 

```python
top_female_ratings = mean_ratings.sort_values(by='F', ascending=False)
top_female_ratings[:10]
```

Suponha que você quisesse saber quais são os filmes que
mais dividem as opiniões entre os telespectadores masculinos e femininos

Vamos adicionar uma coluna "diff"

```python
mean_ratings['diff'] = mean_ratings['M'] - mean_ratings['F']
```

Ordenar por 'diff' resulta nos filmes com as maiores diferenças de avaliação, de modo que podemos ver quais são os preferidos pelas mulheres:

```python
sorted_by_diff = mean_ratings.sort_values(by='diff')
sorted_by_diff[:10]
```

Vamos inverter a ordem das linhas para descobrir os mais preferidos pelos homens e pouco preferido pelas mulheres

```python
sorted_by_diff = sorted_by_diff.dropna()
sorted_by_diff[::-1][:10]
```
